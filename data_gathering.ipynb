# Scraping

```python
# suppose some scraping methods were executed successfully, while others failed
# I want to know what made the process failed
# thus reporting dataframe contains the following columns

df_scraping_result = pd.DataFrame({
	"web_url": pd.Series(dtype="str"), 
	"is_gambling_site": pd.Series(dtype="bool"),
	"is_error": pd.Series(dtype="bool"), 
	"error_scraping_desc": pd.Series(dtype="str"),
	"scraped_elements": pd.Series(dtype="str"),
	"exception_raised": pd.Series(dtype="datetime64[ns]") #for reporting purposes
})
df_scraping_result
```

## Execution

```python
driver = helper.get_webdriver()

df_row_count = df.shape[0]

scraping_result_list = [
    helper.get_scraping_result(web_url, is_gambling_site, driver, item_position=i+1, df_row_count=df_row_count) \
        for i, (web_url, is_gambling_site) in enumerate(zip(df["web_url"], df["is_gambling_site"]))
]

df_scraping_result = pd.concat(
	[df_scraping_result, pd.DataFrame(scraping_result_list)], 
	axis=0,
    ignore_index=True
)

driver.quit()
```

# Export Dataset
